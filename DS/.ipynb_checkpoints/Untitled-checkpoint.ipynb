### Working directory and file path

```python
data_dir = "X:/Richards_flow/largescale/"
file_name = "model_domain_quad.tec"
file_path = os.path.join(data_dir,file_name)
```

### Read file and process it

```python
#read the tec file
size, steps, Headers, D = rdr.readTecfile(file_path)
#Convert the coarse grid in to 51x31 array
df = rdr.Convertetoarray(D, "tec", 51, 31)
for i in range(np.shape(df)[0]):
    print(Headers[i + 3], np.mean(df[i, -1, 0, :]))
    #print(np.mean(df[2, -1, :, :]))
```
