## Reading BRNS files

```python
import os
import numpy as np
import pandas as pd
import data_reader.data_processing as proc
import matplotlib.pyplot as plt

#Required functions:
#Calculate volumetric mean of rate of reaction
def calcvolmeanrate (ratedata, domain):
    firstnode = ratedata.iloc[-100]
    lastnode = ratedata.iloc[-1]        
    sidelength = domain.iloc[-1] - domain.iloc[-2]
    domainvol = domain.iloc[-1] - domain.iloc[-100]
    volmeanrate = ((firstnode + lastnode)*sidelength/2 + sum(ratedata.iloc[-101:-1])*sidelength)/domainvol
    
    return volmeanrate
```

### Directory of interest

```python
datadir = "X:/check/"
```

### Loading variables of interest:

```python
spdict = proc.speciesdict("Saturated") #master dictionary of all species
ratenames = proc.masterrates("Saturated")
```

### Loading file names

```python
Chems = list(t for t,values in spdict.items() if spdict[t]["State"]== "Dissolved")
micros = list(t for t,values in spdict.items() if (spdict[t]["State"]== "Active") and (spdict[t]["Location"]== "Immobile"))
print(Chems)
print(micros)
```

### Visualizing tracer flow

```python
#Loading data and plotting it
tracerfile = "tr1.dat"
tracer_path = os.path.join(datadir, tracerfile)
tracerend = np.loadtxt(tracer_path)[-101:,0]
#tracerbreakthrough = np.loadtxt(tracer_path)[2425:2525,0]
plt.plot(tracerend)
#plt.plot(tracerbreakthrough)
```

### Loading reactive species data

```python
concdata = []
for c in Chems:
    filename = spdict[c]["Var"]+ ".dat"
    path_data = os.path.join(datadir, filename)
    filedata = np.loadtxt(path_data)
    #concdata.append(filedata[2425:2525,0])
    concdata.append(filedata[-101:,0])
chemdf = pd.DataFrame.from_records(concdata).T
chemdf.columns = Chems
```

### Line plots to visualize concentration profile in 1D

```python
for c in Chems:
    plt.plot(chemdf[c], label = c)
plt.yscale("log")
plt.ylim((0,1000))
plt.legend()
```

### Loading microbial species data

```python
microdata = []
for c in micros:
    filename = spdict[c]["Var"]+ ".dat"
    print(filename)
    path_data = os.path.join(datadir, filename)
    filedata = np.loadtxt(path_data)
    microdata.append(filedata[-101:,0])
microdf = pd.DataFrame.from_records(microdata).T
microdf.columns = micros
```

### Line plots to visualize concentration profile in 1D

```python
for c in micros:
    plt.plot(microdf[c], label = spdict[c]["Graphname"])
plt.legend()
#plt.ylim((0,300))
plt.yscale("log")
```
